```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import streamlit as st
import plotly.express as px
import altair as alt
import seaborn as sns
```

```python
df = pd.read_csv('spotify_2023.csv')
```

```python
df.info()
```

```python
df.drop(columns=['in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'mode', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%', 'key', 'in_apple_charts', 'in_apple_playlists'], inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['column_name'] = pd.to_numeric(df['streams'], errors='coerce')
df = df.dropna(subset=['streams'])
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['release']= df['released_year'] + df['released_month']
bpm= df.groupby(['track_name','released_year','in_spotify_charts']).agg({'bpm':'sum'}).reset_index()
streams= df.groupby(['track_name','released_year','in_spotify_charts']).agg({'streams':'sum'}).reset_index()
```

```python
df = df.dropna()
```

```python
plt.scatter(df['bpm'], df['in_spotify_charts'], alpha=0.5)
plt.xlabel('BPM')
plt.ylabel('Number on Spotify charts')
plt.title('Number on spotify charts based on BPM')
plt.show()
```

```python
plt.hist(df['released_year'], bins=30, alpha=0.5, edgecolor='black') 
plt.xlabel('Release Year')
max_year=int(df['released_year'].max())
min_year=int(df['released_year'].min())
plt.xticks(np.arange(min_year, max_year + 1, 10), rotation=45) 
plt.ylabel('Count')
plt.title("Number of Songs Released by Year on Spotify's most streamed")
plt.show()
```

```python
plt.hist(df['released_month'], bins=12, edgecolor='black', alpha=0.7)
plt.xlabel('Release Month')
plt.ylabel('Song Count')
plt.title("Song Releases by Month on Spotify's most streamed")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
